```python
import pandas as pd
import numpy as np
```

```python
#read in the data
df = pd.read_csv('datasets/GasElecPrices.csv')
```

```python
df.head(20)
```

```python
random = (df['ESC'].max() - df['ESC'].min()) / 60

#for each Nan in ESC, replace with last + random
for i in range(len(df)):
    if np.isnan(df['ESC'][i]):
        df['ESC'][i] = df['ESC'][i-1] + random

df.head(20)
```

```python
random = (df['GSC'].max() - df['GSC'].min()) / 60

#for each Nan in GSC, replace with last + random
for i in range(len(df)):
    if np.isnan(df['GSC'][i]):
        df['GSC'][i] = df['GSC'][i-1] + random

df.head(20)
```

```python
#round GSC and ESC columns to 2 decimal places
df['GSC'] = df['GSC'].round(2)
df['ESC'] = df['ESC'].round(2)



#export csv
df.to_csv('datasets/GasElecPrices.csv', index=False)

df.head(20)
```
