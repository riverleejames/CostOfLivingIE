### 2.3.1. Missing Data <a class="anchor" id="MissingData"></a>

In addition to incorrect datatypes, another common problem when dealing with real-world data is missing values. These can arise for many reasons and have to be either filled in or removed before we train a machine learning model. First, let’s get a sense of how many missing values are in each column

While we always want to be careful about removing information, if a column has a high percentage of missing values, then it probably will not be useful to our model. The threshold for removing columns should depend on the problem

```python
def missing_values_table(df):
        mis_val = df.isnull().sum()
        mis_val_percent = 100 * df.isnull().sum() / len(df)
        mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)
        mis_val_table_ren_columns = mis_val_table.rename(
        columns = {0 : 'Missing Values', 1 : '% of Total Values'})
        mis_val_table_ren_columns = mis_val_table_ren_columns[
            mis_val_table_ren_columns.iloc[:,1] != 0].sort_values(
        '% of Total Values', ascending=False).round(1)
        print ("Your selected dataframe has " + str(df.shape[1]) + " columns.\n"      
            "There are " + str(mis_val_table_ren_columns.shape[0]) +
              " columns that have missing values.")
        return mis_val_table_ren_columns
```

```python
# Average weekly hours worked in Ireland: https://www.statista.com/statistics/419584/main-job-average-weekly-working-hours-ireland-y-on-y/
# On average 35 hours per week is worked in Ireland.

#Function that returns a value where if earnings between €352.01 and €424, sub-classes AX and AL, the maximum weekly PRSI Credit of €12 is reduced by one-sixth of earnings in excess of €352.01. 
#https://www.gov.ie/en/collection/06bf07-prsi-contribution-rates-and-user-guide-sw14/
def prsiCredit(earnings):
    if earnings > 352.01 and earnings <= 424:
        return 12 - ((earnings - 352.01) / 6)
    else:
        return 0

def getSalariesAndApplyTaxes(df):
    uscLowestRate = 12012 * 0.005
    # Add weekly hours worked column to minimum wage dataframe
    df['WeeklyHours'] = 35

    # add a column to minimum wage dataframe that calculates the the weekly wage based on the hourly wage and weekly hours worked
    df['GrossWeeklyWage'] = df['Amount'] * df['WeeklyHours']

    # add a column to minimum wage dataframe that calculates the the monthly wage based on the hourly wage and weekly hours worked
    df['GrossMonthlyWage'] = df['GrossWeeklyWage'] * 4

    # add a column to minimum wage dataframe that calculates the the annual wage based on the hourly wage and weekly hours worked
    df['GrossAnnualWage'] = df['GrossWeeklyWage'] * 52

    # Add column PersonalTaxCredit to minimum wage dataframe set it to 1,650 except 2022 and 2023 where it is 1,700
    df['PersonalTaxCredit'] = 1650
    df.loc[df['Year'] == 2022, 'PersonalTaxCredit'] = 1700
    df.loc[df['Year'] >= 2023, 'PersonalTaxCredit'] = 1775

    # Add column EmployeeTaxCredit to minimum wage dataframe set it to 1,650 except 2022 and 2023 where it is 1,700
    df['EmployeeTaxCredit'] = 1650
    df.loc[df['Year'] == 2022, 'EmployeeTaxCredit'] = 1700
    df.loc[df['Year'] >= 2023, 'EmployeeTaxCredit'] = 1775

    # Add tax credit column to minimum wage dataframe sum personal tax credit and employee tax credit
    df['TaxCredit'] = df['PersonalTaxCredit'] + df['EmployeeTaxCredit']

    # set tax rate to 20% for all years
    df['TaxRate'] = 0.2

    # Standard rate of USC
    # Rate	Income band
    # 0.5%	Up to €12,012
    # 2%	From €12,012.01 to €21,295
    # 4.5%	From €21,295.01 to €70,044
    # 8%	From €70,044.01 and over
    # 11%	Self-employed income over €100,000

    uscLowestRate = 12012 * 0.005

    # Add USC payable column to minimum wage dataframe
    df['USCPayable'] = (df['GrossAnnualWage'] - 12012) * 0.02 + uscLowestRate

    # add a prsipayable column and if annual wage is greater than 18304.53 set prsi to GrossAnnualWage * 0.04 else set it to 0
    df['PRSIPayable'] = 0
    df.loc[df['GrossWeeklyWage'] > 352.01, 'PRSIPayable'] = (
        df['GrossAnnualWage'] * 0.04) - (df['GrossWeeklyWage'].apply(prsiCredit) * 52)

    # add total liabilities column to minimum wage dataframe if total liabilities is less than 0 set it to 0
    df['YearlyTax'] = (df['GrossAnnualWage'] * (df['TaxRate']) +
                       (df['USCPayable'] + df['PRSIPayable']) - df['TaxCredit'])
    df.loc[df['YearlyTax'] < 0, 'YearlyTax'] = 0 + \
        df['USCPayable'] + df['PRSIPayable']

    # add net annual wage column to minimum wage dataframe
    df['NetAnnualWage'] = df['GrossAnnualWage'] - df['YearlyTax']

    return df

#Function that returns a value where if earnings between €352.01 and €424, sub-classes AX and AL, the maximum weekly PRSI Credit of €12 is reduced by one-sixth of earnings in excess of €352.01. 
#https://www.gov.ie/en/collection/06bf07-prsi-contribution-rates-and-user-guide-sw14/
def prsiCredit(earnings):
    if earnings > 352.01 and earnings <= 424:
        return 12 - ((earnings - 352.01) / 6)
    else:
        return 0
```

```python
# Create a function that given a dataframe will use national average electricity usage of 4,200Kw/h and national average gas usage of 11,000Kw/h to calculate the Estimated Annual Bills for gas and electricity 

def getEstimatedAnnualBills(df):
    tax = 1.09
    gasUsage = 11000 
    elecUsage = 4200

    #Add a column to the dataframe that calculates the estimated annual gas bill
    #convert cents per unit to euro per unit
    df['GasUnit'] = df['GasUnit'] / 100
    df['EABGAS'] = (df['GasUnit'] * gasUsage) * tax + df['GSC']

    #Add a column to the dataframe that calculates the estimated annual electricity bill
    #convert cents per unit to euro per unit
    df['ElecUnit'] = df['ElecUnit'] / 100
    df['EABELEC'] = (df['ElecUnit'] * elecUsage) * tax + df['ESC']

    #Add a column to the dataframe that calculates the estimated annual bill for both gas and electricity
    df['EAB'] = df['EABGAS'] + df['EABELEC']

    return df
```
